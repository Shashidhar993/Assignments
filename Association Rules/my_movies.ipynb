```python
# Importing libaries mlxtend
!pip install mlxtend
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
# using apriori algorithm
frequent_itemsets = apriori(df1, min_support= 0.1, use_colnames=True)
frequent_itemsets
```

```python
# Using association rules concept
rules = association_rules(frequent_itemsets, metric='lift', min_threshold=0.7)
rules
rules.sort_values('lift',ascending = False).tail()
```

```python
# Sorting the values in descending manner of the lift values
rules.sort_values('lift',ascending = False)[0:20]
```

```python
# Lift values which are more than one
rules[rules.lift>1]
```

```python
plt.rcParams['figure.figsize']=(6,3)
plt.rcParams['figure.dpi']=(200)
```

```python
# Histogram between support and confidence values.
rules[['support','confidence']].hist()
```

```python
# Histogram between the support,confidence and lift values.
rules[['support','confidence','lift']].hist()
```
